```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv("cleaned_data.csv")
```

```python
car = ds[["company", "name", "year", "kms_driven", "fuel_type", "Price"]]
```

```python
ax = sns.histplot(car['year'], kde=True, bins=15, color='purple')
plt.title("Yearwise Count")
plt.xlabel("Year", fontsize=12)
plt.ylabel("Count", fontsize=12)
for container in ax.containers:
    ax.bar_label(container, fmt='%d', label_type='edge', fontsize=10, color='black')
plt.show()
```

```python
sns.boxplot(x='fuel_type', y='Price', data=car, palette='coolwarm')
plt.title("Box Plot")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.swarmplot(x='company', y='Price', data=car, palette='Set3')
plt.title("Swarm Plot")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.countplot(x='company', data=car, palette='pastel')
plt.title("Count Plot")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.pairplot(car, hue='fuel_type')
plt.show()
```

```python
correlation = car.corr()  # Correlation matrix
sns.heatmap(correlation, annot=True, cmap='coolwarm')
plt.title("Heatmap")
plt.show()
```

```python
plt.imshow(car, cmap='Price', interpolation='nearest')
plt.colorbar()
plt.title("Heatmap")
plt.show()
```

```python
car['range'] = 0
```

```python
for i in car.index:
    if car.loc[i, "Price"] < 100000:
        car.loc[i, "range"] = 0
    elif car.loc[i, "Price"] < 200000:
        car.loc[i, "range"] = 1
    elif car.loc[i, "Price"] < 300000:
        car.loc[i, "range"] = 2
    elif car.loc[i, "Price"] < 400000:
        car.loc[i, "range"] = 3
    elif car.loc[i, "Price"] < 500000:
        car.loc[i, "range"] = 4
    else:
        car.loc[i, "range"] = 5    
```

```python
car.groupby("range")["range"].count()
```
