```python
#1. OneHotEncoder
#Encode categorical features as a one-hot numeric array.

# Spain France Germany
# 0      0      1
# 1      0      0

from sklearn.preprocessing import OneHotEncoder
#Sparse = False creates horizontal array(Use when more than 2 values are there)
ohe = OneHotEncoder(sparse=False)
#Convert Country column to one hot array
encoded_data = ohe.fit_transform(ds[['Country']])
#Apply this ohe on dataframe Country column
encoded_df = pd.DataFrame(encoded_data, columns=ohe.get_feature_names_out(['Country']))
#Append new encoded_df to original dataset
ds = pd.concat([ds, encoded_df], axis=1)
#Drop Country column as we got alternative for this
ds = ds.drop(columns=['Country'])
```
